```python
x=Customers.iloc[:,1:].values
```

```python
x.corr()
```

```python
sns.heatmap(x.corr())
plt.show()
```

```python
from sklearn.preprocessing import minmax_scale
x_Minmax= minmax_scale(x)
```

```python
x_Minmax
```

```python
from sklearn.decomposition import PCA
pca=PCA()
pca.fit(x)
```

```python
pca.explained_variance_ratio_
```

```python
#Dropping PCA as there seems not much Corelation between elements
```

```python
from sklearn.cluster import KMeans
WCSS=[]
for i in range (1,11):
            kmeans=KMeans(n_clusters=i,init='k-means++',n_init=10)
            kmeans.fit(x_Minmax)
            WCSS.append(kmeans.inertia_)
```

```python
WCSS
```

```python
plt.plot(range(1,11),WCSS)
plt.show()
```

```python
#From above we are getting number of clusters as 3
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',n_init=10)
kmeans=kmeans.fit(x_Minmax)
y_means=kmeans.labels_
display (y_means)
```

```python
y_means
```

```python
plt.scatter (x[y_means==0,0],x[y_means==0,1],s=100,c='r')
plt.scatter (x[y_means==1,0],x[y_means==1,1],s=100,c='b')
plt.scatter (x[y_means==2,0],x[y_means==2,1],s=100,c='g')

plt.show()
```

```python
Cluster=pd.DataFrame(y_means,columns=['Cluster Number'])
```

```python
Marget_Segmentation=pd.concat([Customers,Cluster],axis=1)
```

```python
Marget_Segmentation
```

```python
Marget_Segmentation.to_excel("Marget_Segmentation.xlsx")
```
