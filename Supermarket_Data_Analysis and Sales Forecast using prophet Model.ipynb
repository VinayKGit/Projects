#Prepare Data to be used for Building Machine learning Model.Remove all unnecessarycolumns and Label Encode

```python
Sales_DS.columns
```

```python
Sales_DS=Sales_DS.drop(['Row ID','Order ID','Customer ID','Customer Name','Postal Code','Product ID'],axis=1)
```

```python
Sales_DS
```

```python
Sales_DS.columns
```

```python
Sales_Project=Sales_DS[['Order Date','Segment','Country', 'City', 'State', 'Region', 'Category', 'Sub-Category',
       'Product Name', 'Sales']]
```

```python
Sales_Project=Sales_Project.sort_values(by='Order Date')
```

```python
Sales_Project
```

```python
Sales_Project['Month']=Sales_Project['Order Date'].dt.month
```

```python
Sales_Project['Year']=Sales_Project['Order Date'].dt.year
```

```python
Sales_Project=Sales_Project.rename(columns={'Order Date':'ds','Sales':'y'})
```

```python
Sales_Project
```

#Forecast sales using prophet Model for next 2 Years

```python
from prophet import Prophet
from prophet.plot import plot_components,plot_plotly
```

```python
model=Prophet()
```

```python
model.fit(Sales_Project)
```

```python
Forecasting=model.make_future_dataframe(periods=730)
```

```python
predictions=model.predict(Forecasting)
```

```python
predictions[['ds','trend','yhat']]
```

```python
plot_plotly(model,predictions)
```

```python
Forecasting=predictions[['ds','trend','yhat']]
```

```python
Forecasting.plot(x='ds',y=['trend','yhat'])
plt.show()
```
