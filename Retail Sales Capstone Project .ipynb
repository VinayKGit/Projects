```python
obj_columns = list(Retail_DF1.select_dtypes(include=['object']).columns)
obj_columns
```

```python
Retail_DF1_onehot= pd.get_dummies(data=Retail_DF1,columns=obj_columns, dtype =int)
Retail_DF1_onehot
```

```python
Retail_DF1_onehot
```

```python
Retail_DF1_onehot=Retail_DF1_onehot.set_index('Date')
```

```python
Retail_DF1_onehot=Retail_DF1_onehot.sort_index()
```

```python
#Since Dataset is Huge Resample to weekly Basis and Later can forecast Yearly and Monthy Sales by resampling
```

```python
Retail_DF1_onehot=Retail_DF1_onehot.resample('w').mean()
```

```python
Retail_DF1_onehot
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
seasonal_decompose(Retail_DF1_onehot['Sales'],period=52).plot()
```

```python
exo=Retail_DF1_onehot.drop(['Sales'],axis=1)
```

```python
exo
```

```python
x=Retail_DF1_onehot['Sales']
```

```python
x
```

```python
x.shape
```

```python
Train=x.iloc[:125]
```

```python
Test=x.iloc[125:]
```

```python
exo_Train=exo.iloc[:125,:]
```

```python
exo_Test=exo.iloc[125:,:]
```

```python
from pmdarima import auto_arima
auto_arima(x,m=7,exogenous=exo,D=1,trace=True)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
Model=SARIMAX(Train,order=(2,0,2),exog=exo_Train,seasonal_order=(0,1,1,7))
predictor=Model.fit()
```

```python

predicted=predictor.predict(start=len(Train),end=len(Test)+len(Train)-1,exog=exo_Test,typ='levels')
```

```python
plt.plot(Test)
plt.plot(predicted)
```

```python
from sklearn.metrics import r2_score
r2_score(Test,predicted)
```
